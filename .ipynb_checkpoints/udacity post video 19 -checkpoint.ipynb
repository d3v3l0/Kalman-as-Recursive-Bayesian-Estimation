```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
def gaussian(x, mu, sig):
    return norm.pdf(x, loc = mu, scale = sig)
    #return np.exp(-np.power((x - mu)/sig, 2.)/2)/(np.sqrt(2.*np.pi)*sig)

def plot_gaussian(mu=0, sig= 1, points = 200):
    N = 4
    x = np.linspace(mu-N*sig, mu+N*sig, points)
    y = gaussian(x, mu, sig)
    plt.plot(x,y)
    plt.show()
    

def plot_gauss(sig = 2):
    mu = 5
    points = 200
    x = np.linspace(2, 8, points)
    y = gaussian(x, mu, sig)
    plt.plot(x,y)
    plt.scatter([3,7], [gaussian(3, mu, sig), gaussian(7, mu, sig)])

def plot_density():
    points = 200
    mu = 3
    sig_range = np.linspace(0.1, 10, points)
    plt.plot(sig_range, gaussian(4.9, mu, sig_range))
    plt.show()

plot_gaussian(mu=5, sig = 0.1)
plot_gauss(sig = 0.25)
plot_gauss(sig = 0.5)
plot_gauss(sig = 1)
plot_gauss(sig = 2)
plot_gauss(sig = 8)
plt.show()
plot_density()
```
